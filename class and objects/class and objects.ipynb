##Class and Objects

```python
class Employee():
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def increase_salary(self,percent):
        self.salary+=(self.salary*(percent/100))
    def display(self):
        print(f"Name:{self.name}")
        print (f"Salary:{int(self.salary)}")
e = Employee("Ali", 50000)
e.increase_salary(10)
e.display()
```

#Parking System

```python
class Vehicle:
    def __init__(self, vehicle_number, vehicle_type):
        self.vehicle_number = vehicle_number
        self.vehicle_type = vehicle_type


class Parking_slot:

    def __init__(self, capacity):
        self.capacity = capacity
        self.parked_vehicle = []

    def park(self, vehicle):
        if len(self.parked_vehicle) < self.capacity:
            self.parked_vehicle.append(vehicle)
            print(f"Vehicle {vehicle.vehicle_number} parked successfully")
        else:
            print("Parking is full")

    def remove(self, vehicle_number):
        for v in self.parked_vehicle:
            if v.vehicle_number == vehicle_number:
                self.parked_vehicle.remove(v)
                print("Vehicle removed:", vehicle_number)
                return
        print("Vehicle not found")
        
    def available_slot(self):
        print(self.capacity-len(self.parked_vehicle))

    def search_vehicle(self,vehicle_number):
        for v in self.parked_vehicle:
            if v.vehicle_number==vehicle_number:
                print("Vehicle Available")
                return
            
        print("Vehicle not found")

v1 = Vehicle("WB01AB1234", "Car")
v2 = Vehicle("WB02XY5678", "Bike")

parking = Parking_slot(2)

parking.park(v1)
parking.park(v2)
parking.available_slot()
parking.remove("WB01AB1234")
parking.available_slot()
parking.search_vehicle("wb02xy5678")
```

#Library Management System

```python
class Library:

    def __init__(self):
        self.book_rack = []

    def add_book(self, book_name):
        self.book_rack.append(book_name)
        print(f"{book_name} added to library")

    def borrow_book(self, book):

        if book in self.book_rack:
            self.book_rack.remove(book)
            print("Book borrowed successfully")

        else:
            print("Book not available")


if __name__ == "__main__":

    lib = Library()

    lib.add_book("Python")
    lib.add_book("Django")

    lib.borrow_book("Python")
```
